```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

## 이미지 파일 열기

opencv로도 열수 있지만, shape 순서가 바뀔 때도 있어서 PIL로 여는 것을 선호

```python
path = 'C:/Users/신상훈/Desktop/새 폴더/dog.jpg'

image_pil = Image.open(path)
image = np.array(image_pil)
```

```python
image.shape # 3차원 3은 색상
```

## 이미지 들여다보기

```python
image.shape
```

```python
np.min(image),np.max(image)
```

## 그래프로 시각화 하기

```python
plt.hist(image.ravel(),256,[0,256])
plt.show()
```

## 그림 나타내기

```python
plt.imshow(image)
plt.show()
```

## 이미지 흑백으로 열기

```python
image_pil = Image.open(path).convert("L")
image_bw = np.array(image_pil)
```

```python
image_bw.shape
```

```python
plt.imshow(image_bw,'gray')
plt.show()
```

## 이미지 설정

```python
plt.figure(figsize=(10,5))
plt.imshow(image_bw,'jet')
plt.colorbar()
plt.show()
```

## 이미지에 제목 추가

```python
plt.figure(figsize=(10,5))
plt.imshow(image_bw,'jet')
plt.colorbar()
plt.title('dog')
plt.show()
```
